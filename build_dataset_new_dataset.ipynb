```python
all_group = {}
grps = []
df_sub = df.iloc[:, 0:6]
df_sub.rename(columns = {"Unnamed: 5": "n_de_reporte"}, inplace = True)
for group in df_sub.groupby(by = "Marca").__iter__():
	all_group[group[0].lower()] = []
	for n, row in group[1].iterrows():
		all_group[group[0].lower()].append({row.n_de_reporte:row.apply(lambda x: x.lower() if type(x) == str else x).values.tolist()})
```
